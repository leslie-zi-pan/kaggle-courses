## Selecting data for prediction
 Lets first drop rows (axis=0) where we have nan values

```python
melb_housing_data = melb_housing_data.dropna(axis=0)
```

### Selecting The Prediction Target

Dot notation allows use to pull single column (series) from the dataframe. 

We can also use bracket notation where syntax is not allow in a dot notation (e.g. dataframe.1column)

Lets select the price as the Y (prediction) value

```python
y = melb_housing_data.Price
```

### Choosing "Features"

Data which are used to train and preict the labels are called features. You may have to engineer and either use all or a subset of the features to optimise prediction and training - Less important features may hinder performance of the model. 

```python
# Lets select a few columns for features
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
```

```python
X = melb_housing_data[melbourne_features]
```

Lets visualise the training/val data before continuing


```python
X.describe()
```

```python
X.head()
```

## Building your model

We will use the scikit-learn library to build our models. 

These are the steps to building and using a model: 
- Define the model (decision tree, lasso, etc)... does these include other parameters?
- Fit the model to the data
- Predict the values based on the the data
- Evaluate Determine how well the model is performing. 

```python
from sklearn.tree import DecisionTreeRegressor

# Define the model and fix random state
melbourne_model = DecisionTreeRegressor(random_state=1)

# Fit the model 
split = int(len(y) * 0.7)
melbourne_model.fit(X[:split], y[:split])
```

### Predicting using the model

```python
y_pred = melbourne_model.predict(X[split:])
```

```python
from sklearn.metrics import r2_score, mean_squared_error

mse = mean_squared_error(y[split:], y_pred)
r2 = r2_score(y[split:], y_pred)

print(mse)
print(r2)
```
